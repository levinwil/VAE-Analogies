```python
%load_ext autoreload
%autoreload 2
import sys 
sys.path.append('../scripts')
from model import VAE
from utils import load_mnist, load_birds
import matplotlib.pyplot as plt
import numpy as np
```

```python
X_train, X_test, y_train, y_test = load_mnist()
```

```python
i = 1
X_train = X_train[np.where(y_train <= i)[0]]
y_train = y_train[np.where(y_train <= i)[0]]
X_test = X_test[np.where(y_test <= i)[0]]
y_test = y_test[np.where(y_test <= i)[0]]
```

```python
vae = VAE(input_shape = X_train.shape[1:], 
          std = 2,
          filters = 64,
          lr = 3e-4,
          latent_dim = 64)
```

```python
vae.train(X_train, epochs = 400)
```

```python
vae.load_weights('../weights/best_mnist.hdf5')
```
